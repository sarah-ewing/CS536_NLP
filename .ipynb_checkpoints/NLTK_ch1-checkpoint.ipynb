```python
## lets make some functions
def lexical_diversity(text):
    return(len(set(text))/len(text))

def precentage(count, total):
    return 100 * count / total
```

```python
print(lexical_diversity(text3),
lexical_diversity(text5),
precentage(4, 5))
```
